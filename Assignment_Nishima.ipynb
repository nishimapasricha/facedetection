```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
face_cascade=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
smile_cascade = cv2.CascadeClassifier('haarcascade_smile.xml') 
```

# Face Detection

```python
#reading image
img=cv2.imread("t.jpg",1)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
faces=face_cascade.detectMultiScale(img,scaleFactor=1.05,minNeighbors=5)
#Draw rectangle
for x,y,w,h in faces:
    img=cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,255),2)
plt.imshow(img)
plt.show()    
```

# Eyes Detection

```python
#reading image
img=cv2.imread("d.jpg",1)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
eyes = eye_cascade.detectMultiScale(img,scaleFactor=1.5,minNeighbors=2) 
#To draw a rectangle in eyes 
for (ex,ey,ew,eh) in eyes: 
    img=cv2.rectangle(img,(ex,ey),(ex+ew,ey+eh),(0,127,255),1) 
plt.imshow(img)
plt.show()    
```

# Smile Detection

```python
img=cv2.imread("d.jpg",1)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
faces=face_cascade.detectMultiScale(img,scaleFactor=1.05,minNeighbors=5)
smile = smile_cascade.detectMultiScale(img,scaleFactor=1.5,minNeighbors=25) 
for (ex,ey,ew,eh) in smile: 
            img=cv2.rectangle(img,(ex,ey),(ex+ew,ey+eh),(0,127,255),1) 
plt.imshow(img)
plt.show()    
```

# face and Eyes detection on Web cam

```python
#reading from camera
cap = cv2.VideoCapture(0)  
while 1:   
    r,img = cap.read()  
    faces = face_cascade.detectMultiScale(img, 1.3, 5) 
    for (x,y,w,h) in faces:  
        img=cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,0),2)  
        eyes = eye_cascade.detectMultiScale(img,minNeighbors=10)  
        for (ex,ey,ew,eh) in eyes: 
            cv2.rectangle(img,(ex,ey),(ex+ew,ey+eh),(0,0,0),2) 
    cv2.imshow('img',img) 
    cv2.waitKey(30) & 0xff 
    k = cv2.waitKey(30) & 0xff
    if k == 27: 
        break
cap.release() 
cv2.destroyAllWindows()  
```
